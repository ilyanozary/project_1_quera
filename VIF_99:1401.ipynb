```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["Feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data

vif_data = calculate_vif(pd.DataFrame(X_varR99))
print("Initial VIF values:")
print(vif_data)

def feature_selection_by_vif(X, threshold=6.0):
    high_vif = True
    while high_vif:
        vif_data = calculate_vif(X)
        max_vif = vif_data['VIF'].max()
        if max_vif > threshold:
            max_vif_feature = vif_data.loc[vif_data['VIF'].idxmax(), 'Feature']
            print(f'Removing feature: {max_vif_feature} with VIF: {max_vif}')
            X = X.drop(columns=[max_vif_feature])
        else:
            high_vif = False
    return X, vif_data


XR99_selected, final_vif_dataR99 = feature_selection_by_vif(pd.DataFrame(X_varR99))

print("\nSelected features after VIF reduction:")
print(XR99_selected.columns)

print("\nFinal VIF values:")
print(final_vif_dataR99)
```
